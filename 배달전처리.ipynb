## 1. 파일 불러오기

```python
import pandas as pd
import numpy as np
```

```python
covid = pd.read_csv("KGU_3rd_ORIGIN_KGUDSTNORDER_20210101000000.csv",header=None)
covid
```

#### - 칼럼명 변경

```python
covid.rename(columns={0:'날짜', 1:'시간', 2:'분류', 3:'도', 4:'구', 5:'건수'}, inplace=True)
```

```python
covid
```

##  2. 데이터 전처리

### 1) 서울특별시만 추출

```python
seoul = covid[covid['도'] == '서울특별시']
```

```python
seoul
```

### 2) 필요한 칼럼만 추출; 날짜, 도, 구, 건수

```python
seoul = seoul.drop(['분류','도','시간'], axis=1)
```

```python
seoul
```

### 3) 데이터 정렬; 날짜와 구를 기준으로

```python
seoul = seoul.sort_values(by=['날짜','구'], ascending=True)
seoul = seoul.reset_index()
seoul
```

```python
seoul = seoul.drop('index', axis=1)
seoul
```

### 4) 새로운 리스트 생성; 날짜, 구

#### - 데이터 타입확인

```python
seoul.dtypes
```

#### - 구명 리스트 생성

```python
gu = []
gu = seoul['구'].unique()
gu
```

#### - 날짜 리스트 생성

```python
date = []
date = seoul['날짜'].unique()
len(date)
```

#### - 일별에 따른 구명 합치기

```python
date_gu = []
for i in date:
    for j in gu:
        date_gu.append([i,j])
```

```python
date_gu
```

#### - 데이터 프레임으로 저장; seoul 데이터와 합치기 위함

```python
vv = pd.DataFrame(date_gu)
vv.rename(columns={0:'날짜', 1:'구'}, inplace=True)
```

```python
vv
```

### 5) 최종 데이터; merge 시키기 

```python
fin = pd.merge(vv, seoul, on=['날짜', '구'], how='left')
fin
```

#### - 결측치는 0으로 처리

```python
fin.fillna(0, inplace=True)
```

#### - 건수 데이터 타입을 int로 변환

```python
fin['건수'] = fin['건수'].astype(int)
fin
```

#### - 날짜별 구에 따른 배달건수 합계 추출

```python
fin = fin.groupby(['날짜', '구']).sum()
fin
```

## 3. 최종 데이터 파일로 저장

```python
fin.to_csv("delivery.csv", encoding='cp949')
```
